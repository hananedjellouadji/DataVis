```python
import pandas as pd
import matplotlib.pyplot as plt
import mlxtend
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
ds=pd.read_csv('Market_Basket_Optimisation.csv')
ds
```

```python
te=TransactionEncoder()
te_ary=te.fit(ds).transform(ds)
df=pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

df_sets = df.applymap(encode_units)
frequent_itemsets = apriori(df_sets, min_support=0.001, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.23)
rules
```

```python
rules=association_rules(frequent_itemsets,metric='confidence',min_threshold=0.7)
rules
```
